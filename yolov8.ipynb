```python
import os
import random
import shutil

data_dir = '/content/drive/MyDrive/CSE472/data/images'
label_dir = '/content/drive/MyDrive/CSE472/data/annotations'

train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

# Load the image and label file paths
image_files = [f for f in os.listdir(data_dir) if f.endswith('.png')]
label_files = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

# Make sure there is a label file for each image
image_files = [f for f in image_files if os.path.splitext(f)[0] + '.txt' in label_files]

# Shuffle the list of image files randomly
random.shuffle(image_files)

# Split the image files into train, validation and test sets
train_size = int(len(image_files) * train_ratio)
val_size = int(len(image_files) * val_ratio)
test_size = len(image_files) - train_size - val_size

train_files = image_files[:train_size]
val_files = image_files[train_size:train_size+val_size]
test_files = image_files[train_size+val_size:]

# Create the train, validation and test directories
train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'val')
test_dir = os.path.join(data_dir, 'test')

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Copy the image files to the corresponding directory
def copy_files(files, src_dir, dst_dir):
    for f in files:
        shutil.copy(os.path.join(src_dir, f), os.path.join(dst_dir, f))

copy_files(train_files, data_dir, train_dir)
copy_files(val_files, data_dir, val_dir)
copy_files(test_files, data_dir, test_dir)

# Create the train, validation and test label files
def create_label_files(files, src_dir, dst_dir):
    for f in files:
        label_file = os.path.splitext(f)[0] + '.txt'
        shutil.copy(os.path.join(label_dir, label_file), os.path.join(dst_dir, label_file))

create_label_files(train_files, label_dir, train_dir)
create_label_files(val_files, label_dir, val_dir)
create_label_files(test_files, label_dir, test_dir)


# Define the paths to the input files and output directories
data_dir = '/content/drive/MyDrive/CSE472/data/images/train'
image_dir = '/content/drive/MyDrive/CSE472/data/train/images'
label_dir = '/content/drive/MyDrive/CSE472/data/train/labels'

# Create the output directories if they do not exist
for directory in [image_dir, label_dir]:
    if not os.path.exists(directory):
        os.makedirs(directory)

# Loop through the files in the input directory and move them to the appropriate output directory
for filename in os.listdir(data_dir):
    if filename.endswith('.png'):
        shutil.move(os.path.join(data_dir, filename), os.path.join(image_dir, filename))
    elif filename.endswith('.txt'):
        shutil.move(os.path.join(data_dir, filename), os.path.join(label_dir, filename))

data_dir = '/content/drive/MyDrive/CSE472/data/images/test'
image_dir = '/content/drive/MyDrive/CSE472/data/test/images'
label_dir = '/content/drive/MyDrive/CSE472/data/test/labels'

# Create the output directories if they do not exist
for directory in [image_dir, label_dir]:
    if not os.path.exists(directory):
        os.makedirs(directory)

# Loop through the files in the input directory and move them to the appropriate output directory
for filename in os.listdir(data_dir):
    if filename.endswith('.png'):
        shutil.move(os.path.join(data_dir, filename), os.path.join(image_dir, filename))
    elif filename.endswith('.txt'):
        shutil.move(os.path.join(data_dir, filename), os.path.join(label_dir, filename))

data_dir = '/content/drive/MyDrive/CSE472/data/images/val'
image_dir = '/content/drive/MyDrive/CSE472/data/val/images'
label_dir = '/content/drive/MyDrive/CSE472/data/val/labels'

# Create the output directories if they do not exist
for directory in [image_dir, label_dir]:
    if not os.path.exists(directory):
        os.makedirs(directory)

# Loop through the files in the input directory and move them to the appropriate output directory
for filename in os.listdir(data_dir):
    if filename.endswith('.png'):
        shutil.move(os.path.join(data_dir, filename), os.path.join(image_dir, filename))
    elif filename.endswith('.txt'):
        shutil.move(os.path.join(data_dir, filename), os.path.join(label_dir, filename))
```
